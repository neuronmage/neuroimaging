# Model Learn Derivative

### Model Derivative -- Compares derivative bins based on the derivative of the derivative curve

```python
from glob import glob
from scipy import stats
from math import sqrt
import nibabel as nb
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_palette('muted')


sids = ['WMAZE_001', 'WMAZE_002', 'WMAZE_004', 'WMAZE_005', 'WMAZE_006',
        'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 'WMAZE_012',
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021',
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']


proj_dir = '/home/data/madlab/data/mri/wmaze' 
mask_filenames = []
cope_files = []
for SID in sids:
    mask_filenames_glob = glob(proj_dir + '/roi_analysis/MRthesis/fb4c_2/mask/anat_masks/_subject_id_' 
                               + SID + '/_anatmask_xfm*/*')
    mask_filenames.append(mask_filenames_glob)
    subjcopes_glob = glob(proj_dir + '/scndlvl/wmaze_MRthesis/fixed_before_conditional/model_deriv/' 
                          + SID + '/fixedfx/cope_deriv*')
    subjcopes_glob.sort()
    cope_files.append(subjcopes_glob)
    
    if len(cope_files[-1]) == 0:
        print(SID)
```

## Cell to double-check the array indexing

```python
'''
print '## left hippocampus:', mask_filenames[0][0]
print '## left caudate:', mask_filenames[0][1]
print '## right caudate:', mask_filenames[0][2]
print '## left dmPFC:', mask_filenames[0][3]
print '## right hippocampus:', mask_filenames[0][4]
print '## right dmPFC:', mask_filenames[0][5]
print '## left putamen', mask_filenames[0][6]
print '## right_putamen', mask_filenames[0][7]
'''

print '-----'
print '## Deriv_Bin #0:', cope_files[0][0]
print '## Deriv_Bin #1:', cope_files[0][1]
print '## Deriv_Bin #2:', cope_files[0][2]
print '## Deriv_Bin #3:', cope_files[0][3]
```

### Use binarized mask to obtain activation in left & right hemisphere for each region of interest

```python
all_data = {'subjid':[],
            #HPC
            'lhhp_derivative_bin0':[], 'rhhp_derivative_bin0':[],
            'lhhp_derivative_bin1':[], 'rhhp_derivative_bin1':[], 
            'lhhp_derivative_bin2':[], 'rhhp_derivative_bin2':[],
            'lhhp_derivative_bin3':[], 'rhhp_derivative_bin3':[],
            #Caudate
            'lhcaud_derivative_bin0':[], 'rhcaud_derivative_bin0':[],
            'lhcaud_derivative_bin1':[], 'rhcaud_derivative_bin1':[], 
            'lhcaud_derivative_bin2':[], 'rhcaud_derivative_bin2':[],
            'lhcaud_derivative_bin3':[], 'rhcaud_derivative_bin3':[],
            #Putamen
            'lhput_derivative_bin0':[], 'rhput_derivative_bin0':[],
            'lhput_derivative_bin1':[], 'rhput_derivative_bin1':[], 
            'lhput_derivative_bin2':[], 'rhput_derivative_bin2':[],
            'lhput_derivative_bin3':[], 'rhput_derivative_bin3':[]}

for i in range(len(sids)):
    all_data['subjid'].append(sids[i])
    lh_hp_img = nb.load(mask_filenames[i][0])
    rh_hp_img = nb.load(mask_filenames[i][4])
    lh_caud_img = nb.load(mask_filenames[i][1])
    rh_caud_img = nb.load(mask_filenames[i][2])
    lh_put_img = nb.load(mask_filenames[i][6])
    rh_put_img = nb.load(mask_filenames[i][7])
    
    derivative_bin0_img = nb.load(cope_files[i][0])
    derivative_bin1_img = nb.load(cope_files[i][1])
    derivative_bin2_img = nb.load(cope_files[i][2])
    derivative_bin3_img = nb.load(cope_files[i][3])
    
    
    ########################
    ### Left Hippocampus ###
    ########################
    
    # derivative_bin0
    lh_hp_derivative_bin0_data = derivative_bin0_img.get_data()[lh_hp_img.get_data() > 0.]
    all_data['lhhp_derivative_bin0'].append(np.mean(lh_hp_derivative_bin0_data))
    
    # derivative_bin1
    lh_hp_derivative_bin1_data = derivative_bin1_img.get_data()[lh_hp_img.get_data() > 0.]
    all_data['lhhp_derivative_bin1'].append(np.mean(lh_hp_derivative_bin1_data))
    
    # derivative_bin2
    lh_hp_derivative_bin2_data = derivative_bin2_img.get_data()[lh_hp_img.get_data() > 0.]
    all_data['lhhp_derivative_bin2'].append(np.mean(lh_hp_derivative_bin2_data))
    
    # derivative_bin3
    lh_hp_derivative_bin3_data = derivative_bin3_img.get_data()[lh_hp_img.get_data() > 0.]
    all_data['lhhp_derivative_bin3'].append(np.mean(lh_hp_derivative_bin3_data))
    
    
    #########################
    ### Right Hippocampus ###
    #########################
    
    # derivative_bin0
    rh_hp_derivative_bin0_data = derivative_bin0_img.get_data()[rh_hp_img.get_data() > 0.]
    all_data['rhhp_derivative_bin0'].append(np.mean(rh_hp_derivative_bin0_data))
    
    # derivative_bin1
    rh_hp_derivative_bin1_data = derivative_bin1_img.get_data()[rh_hp_img.get_data() > 0.]
    all_data['rhhp_derivative_bin1'].append(np.mean(rh_hp_derivative_bin1_data))
    
    # derivative_bin2
    rh_hp_derivative_bin2_data = derivative_bin2_img.get_data()[rh_hp_img.get_data() > 0.]
    all_data['rhhp_derivative_bin2'].append(np.mean(rh_hp_derivative_bin2_data))
    
    # derivative_bin3
    rh_hp_derivative_bin3_data = derivative_bin3_img.get_data()[rh_hp_img.get_data() > 0.]
    all_data['rhhp_derivative_bin3'].append(np.mean(rh_hp_derivative_bin3_data))

  
    ####################
    ### Left Caudate ###
    ####################
    
    # derivative_bin0
    lh_caud_derivative_bin0_data = derivative_bin0_img.get_data()[lh_caud_img.get_data() > 0.]
    all_data['lhcaud_derivative_bin0'].append(np.mean(lh_caud_derivative_bin0_data)) 
    
    # derivative_bin1
    lh_caud_derivative_bin1_data = derivative_bin1_img.get_data()[lh_caud_img.get_data() > 0.]
    all_data['lhcaud_derivative_bin1'].append(np.mean(lh_caud_derivative_bin1_data))
    
    # derivative_bin2
    lh_caud_derivative_bin2_data = derivative_bin2_img.get_data()[lh_caud_img.get_data() > 0.]
    all_data['lhcaud_derivative_bin2'].append(np.mean(lh_caud_derivative_bin2_data))
    
    # derivative_bin3
    lh_caud_derivative_bin3_data = derivative_bin3_img.get_data()[lh_caud_img.get_data() > 0.]
    all_data['lhcaud_derivative_bin3'].append(np.mean(lh_caud_derivative_bin3_data)) 
    
    
    #####################
    ### Right Caudate ###
    ##################### 
    
    # derivative_bin0
    rh_caud_derivative_bin0_data = derivative_bin0_img.get_data()[rh_caud_img.get_data() > 0.]
    all_data['rhcaud_derivative_bin0'].append(np.mean(rh_caud_derivative_bin0_data))

    # derivative_bin1
    rh_caud_derivative_bin1_data = derivative_bin1_img.get_data()[rh_caud_img.get_data() > 0.]
    all_data['rhcaud_derivative_bin1'].append(np.mean(rh_caud_derivative_bin1_data))
    
    # derivative_bin2
    rh_caud_derivative_bin2_data = derivative_bin2_img.get_data()[rh_caud_img.get_data() > 0.]
    all_data['rhcaud_derivative_bin2'].append(np.mean(rh_caud_derivative_bin2_data))
    
    # derivative_bin3
    rh_caud_derivative_bin3_data = derivative_bin3_img.get_data()[rh_caud_img.get_data() > 0.]
    all_data['rhcaud_derivative_bin3'].append(np.mean(rh_caud_derivative_bin3_data))
    
    
    ####################
    ### Left Putamen ###
    ####################
    
    # derivative_bin0
    lh_put_derivative_bin0_data = derivative_bin0_img.get_data()[lh_put_img.get_data() > 0.]
    all_data['lhput_derivative_bin0'].append(np.mean(lh_put_derivative_bin0_data))
    
    # derivative_bin1
    lh_put_derivative_bin1_data = derivative_bin1_img.get_data()[lh_put_img.get_data() > 0.]
    all_data['lhput_derivative_bin1'].append(np.mean(lh_put_derivative_bin1_data))
    
    # derivative_bin2
    lh_put_derivative_bin2_data = derivative_bin2_img.get_data()[lh_put_img.get_data() > 0.]
    all_data['lhput_derivative_bin2'].append(np.mean(lh_put_derivative_bin2_data))
    
    # derivative_bin3
    lh_put_derivative_bin3_data = derivative_bin3_img.get_data()[lh_put_img.get_data() > 0.]
    all_data['lhput_derivative_bin3'].append(np.mean(lh_put_derivative_bin3_data))
    
    
    #####################
    ### Right Putamen ###
    #####################
    
    # derivative_bin0
    rh_put_derivative_bin0_data = derivative_bin0_img.get_data()[rh_put_img.get_data() > 0.]
    all_data['rhput_derivative_bin0'].append(np.mean(rh_put_derivative_bin0_data)) 

    # derivative_bin1
    rh_put_derivative_bin1_data = derivative_bin1_img.get_data()[rh_put_img.get_data() > 0.]
    all_data['rhput_derivative_bin1'].append(np.mean(rh_put_derivative_bin1_data))
    
    # derivative_bin2
    rh_put_derivative_bin2_data = derivative_bin2_img.get_data()[rh_put_img.get_data() > 0.]
    all_data['rhput_derivative_bin2'].append(np.mean(rh_put_derivative_bin2_data))
    
    # derivative_bin3
    rh_put_derivative_bin3_data = derivative_bin3_img.get_data()[rh_put_img.get_data() > 0.]
    all_data['rhput_derivative_bin3'].append(np.mean(rh_put_derivative_bin3_data)) 
    
    
all_data_df = pd.DataFrame(all_data)
```

### Combine the two hemispheres to create one regional average

```python
all_data_df['hp_all_derivative_bin0'] = (all_data_df['lhhp_derivative_bin0']
                                         + all_data_df['rhhp_derivative_bin0'])/2.
all_data_df['hp_all_derivative_bin1'] = (all_data_df['lhhp_derivative_bin1'] 
                                         + all_data_df['rhhp_derivative_bin1'])/2.
all_data_df['hp_all_derivative_bin2'] = (all_data_df['lhhp_derivative_bin2']
                                         + all_data_df['rhhp_derivative_bin2'])/2.
all_data_df['hp_all_derivative_bin3'] = (all_data_df['lhhp_derivative_bin3']
                                         + all_data_df['rhhp_derivative_bin3'])/2.


all_data_df['caud_all_derivative_bin0'] = (all_data_df['lhcaud_derivative_bin0'] 
                                           + all_data_df['rhcaud_derivative_bin0'])/2.
all_data_df['caud_all_derivative_bin1'] = (all_data_df['lhcaud_derivative_bin1'] 
                                           + all_data_df['rhcaud_derivative_bin1'])/2.
all_data_df['caud_all_derivative_bin2'] = (all_data_df['lhcaud_derivative_bin2']
                                           + all_data_df['rhcaud_derivative_bin2'])/2.
all_data_df['caud_all_derivative_bin3'] = (all_data_df['lhcaud_derivative_bin3'] 
                                           + all_data_df['rhcaud_derivative_bin3'])/2.


all_data_df['put_all_derivative_bin0'] = (all_data_df['lhput_derivative_bin0'] 
                                          + all_data_df['rhput_derivative_bin0'])/2.
all_data_df['put_all_derivative_bin1'] = (all_data_df['lhput_derivative_bin1']
                                          + all_data_df['rhput_derivative_bin1'])/2.
all_data_df['put_all_derivative_bin2'] = (all_data_df['lhput_derivative_bin2'] 
                                          + all_data_df['rhput_derivative_bin2'])/2.
all_data_df['put_all_derivative_bin3'] = (all_data_df['lhput_derivative_bin3'] 
                                          + all_data_df['rhput_derivative_bin3'])/2.
```

```python
hpc_chart = {}
hpc_chart['0_HPC-derivative_bin0'] = all_data_df['hp_all_derivative_bin0']
hpc_chart['1_HPC-derivative_bin1'] = all_data_df['hp_all_derivative_bin1'] 
hpc_chart['2_HPC-derivative_bin2'] = all_data_df['hp_all_derivative_bin2'] 
hpc_chart['3_HPC-derivative_bin3'] = all_data_df['hp_all_derivative_bin3']
hpc_chart_df = pd.DataFrame.from_dict(hpc_chart)
```

## Hippocampus Activation

```python
print 'bin0 mean:', np.mean(all_data_df['hp_all_derivative_bin0'])
print 'bin0 std:', np.std(all_data_df['hp_all_derivative_bin0'])
print ""
print 'bin1 mean:', np.mean(all_data_df['hp_all_derivative_bin1'])
print 'bin1 std:', np.std(all_data_df['hp_all_derivative_bin1'])
print ""
print 'bin2 mean:', np.mean(all_data_df['hp_all_derivative_bin2'])
print 'bin2 std:', np.std(all_data_df['hp_all_derivative_bin2'])
print ""
print 'bin3 mean:', np.mean(all_data_df['hp_all_derivative_bin3'])
print 'bin3 std:', np.std(all_data_df['hp_all_derivative_bin3'])
print ""


N = 4
conditions = ['derivative Bin0', 'derivative Bin1', 'derivative Bin2', 'derivative Bin3']
hp_allsubjs = [all_data_df['hp_all_derivative_bin0'], 
               all_data_df['hp_all_derivative_bin1'],
               all_data_df['hp_all_derivative_bin2'],
               all_data_df['hp_all_derivative_bin3']]
ind = np.arange(N)
width = 0.5
fig, ax = plt.subplots(figsize = (10,5))
ax0 = sns.boxplot(data = hp_allsubjs, color = "#278fea", width = 0.3)
ax2 = sns.swarmplot(data = hp_allsubjs, color='.25')
ax.set_xticks(ind)
ax.set_xticklabels(conditions)
ax.set_ylabel("Arbitrary units")
ax.set_title("HPC Activation")
plt.show()
```
