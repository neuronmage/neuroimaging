# Model RSA Learning Curve + PMA
## Isolating trials of learning -- positive derivative value, negative derivative value
## Parametric Modulation -- add event and mod EVs

```python
from glob import glob
from pylab import *
import shutil 
import os
from os.path import join, split, basename
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette('muted')

subsX = ['WMAZE_001'] #test sub
subs = ['WMAZE_001', 'WMAZE_002', 'WMAZE_004', 'WMAZE_005', 'WMAZE_006',
        'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 'WMAZE_012',
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021',
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']

database = {'s1r1_learn':[], 's1r1_nonlearn':[], 's1r2_learn':[], 's1r2_nonlearn':[], #dict to hold data during iteration
            's2r1_learn':[], 's2r1_nonlearn':[], 's2r2_learn':[], 's2r2_nonlearn':[],
            's3r1_learn':[], 's3r1_nonlearn':[], 's3r2_learn':[], 's3r2_nonlearn':[]}

for sub in subs:
    #print(sub)
    sub_dir = '/home/data/madlab/Mattfeld_WMAZE/'
    ev_dir = join(sub_dir, 'behav/{0}/model_RSA4/'.format(sub))
    behav_runs = sorted(glob(join(sub_dir, 'behav/{0}/{0}_wmazebl_2015*.txt'.format(sub))))
    learning_files = sorted(glob(join(sub_dir,'behav/{0}/B_pmode_set*.txt'.format(sub))))  
   
    if os.path.exists(ev_dir): #removes existing EV files if they exist
        globber = glob(join(ev_dir, '*.txt')) #get everything in folder
        for x in globber: #iterate through folder and delete
            os.remove(x)
    else: #make dir if nonexistant 
         os.makedirs(ev_dir)
            
    for i in range(3): #3 stimulus sets
        behav_run1 = pd.read_table(behav_runs[i*2]) #load behavioral files
        behav_run2 = pd.read_table(behav_runs[i*2+1])
        
        for j in range(2): #2 runs per set    
            learning_curve = np.loadtxt(learning_files[i]) #Smith et al. (2004) learning curve algorithm
            deriv_file = (np.gradient(learning_curve)*10)[1:] #learning curve derivative 
            if j == 0: #curve and derivative are by set -- must split
                learning_curve = learning_curve[:40]
                deriv_file = deriv_file[:40]
            else:
                learning_curve = learning_curve[40:80]
                deriv_file = deriv_file[40:80]           
            
            behav_os = eval('behav_run{0}["StimOnset"].values'.format(j+1)) #onset time
            behav_resp = eval('behav_run{0}["Resp"].values'.format(j+1)) #Ss response  
            behav_type = eval('behav_run{0}["TrialType"].values'.format(j+1)) #Type of trial [A,B,C,BL]
            trial_shift = behav_type[:-1] #duplicate behavior column and remove last index
            trial_shift = np.insert(trial_shift, 0, -1) #insert dummy value into first index
            trial_shift2 = behav_type[1:] #shift TrialType back and insert dummy (-1) in last index
            trial_shift2 = np.append(trial_shift2, -1)

            b_indices = np.where((behav_type == 'B'))[0] #indices of all Bs in dataset without preceeding BL trials
            b_BL_indices = np.where((behav_type == 'B') & (trial_shift == 'BL'))[0] #B trials with preceeding BLs 
            all_remaining = np.where((trial_shift2!='B') | (behav_type=='BL'))[0] #BL trials or fixed without Bs               

            #identify bad Bs for removal in learning curve and derivative files
            bad_Bs = [] 
            bad_Bs.extend(b_BL_indices) #iterates through and adds Bs with preceeding BLs to bad_Bs 
            for curr_B in b_indices: #iterates through all Bs without preceeding BLs
                if not curr_B in bad_Bs: #if the current B is not in bad Bs              
                    #if behav_resp[curr_B] == 'NR': #identify non-response B trials
                        #bad_Bs.append(curr_B) #add to bad_Bs if nonresponse trial                 
                    if curr_B in [0, 157, 158, 159, 160, 317, 318, 319]: #indices if B trial comes first (WMAZE_001 run 6)  
                        bad_Bs.append(curr_B) #add to bad_Bs if B is first trial in run
            bad_Bs.sort() #organize by index       
            bad_B_ind = sorted([k for k, f in enumerate(b_indices) if f in bad_Bs]) #indices for bad Bs within group of Bs      
            bad_B_ind = bad_B_ind[::-1] #reverse order of Bs to be removed   

            #DERIV FILES   
            temp1 = list(deriv_file) #temp version of deriv file as list
            for curr_bad_B in bad_B_ind: #iterate through bad_B indices
                temp1.pop(curr_bad_B) #remove any deriv values associated with Bad Bs  
            new_deriv = np.array(temp1) #save without removed Bs
            learn_derivs = np.where(new_deriv > 0)[0] #any indices where deriv value is greater than 0
            nonlearn_derivs = np.where(new_deriv <= 0)[0] #any indices where deriv value is equal to or less than 0
            database['s{0}r{1}_learn'.format(i+1,j+1)].append(len(learn_derivs))
            database['s{0}r{1}_nonlearn'.format(i+1,j+1)].append(len(nonlearn_derivs))

            #removes bad B INDICES from B-list
            temp2 = list(b_indices) #temp version of b_indices as list
            for curr_bad_B in bad_B_ind: #iterate through bad_B indices
                temp2.pop(curr_bad_B) #remove any Bs at bad_Bs indices
            good_B_ind = np.array(temp2) #save without removed Bs  
            
            f_learn = good_B_ind[learn_derivs] #grabs the B trials with a positive value derivative
            f_nonlearn = good_B_ind[nonlearn_derivs] 
            f_learn = [x - 1 for x in f_learn] #iterates and gets indices for fixed before Bs
            f_nonlearn = [x - 1 for x in f_nonlearn]  
                       
            #learning curve values at selected indices  
            temp3 = np.array(learning_curve) #temp version of learning_curve as list
            learn_values = temp3[learn_derivs]
            nonlearn_values = temp3[nonlearn_derivs]

            #onsets for fixed before respective conditional         
            learn_onsets = behav_os[f_learn] #use indices to isolate values
            nonlearn_onsets = behav_os[f_nonlearn]
            remaining_onsets = behav_os[all_remaining[:-3]] 
               
            mtrx = np.vstack((behav_os, np.ones(len(behav_os))*2.5, #create 3 column matrix [onsets, durations, amplitudes]
                              np.ones(len(behav_os)))).T 
            learn_event_mtrx = np.vstack((learn_onsets, np.ones(len(learn_onsets))*2.5, 
                                    np.ones(len(learn_onsets)))).T #constant amplitude for event matrix
            nonlearn_event_mtrx = np.vstack((nonlearn_onsets, np.ones(len(nonlearn_onsets))*2.5, 
                                       np.ones(len(nonlearn_onsets)))).T
            learn_mod_mtrx = np.vstack((learn_onsets, np.ones(len(learn_onsets))*2.5, 
                                        learn_values)).T #amplitude is modulated by learning curve
            nonlearn_mod_mtrx = np.vstack((nonlearn_onsets, np.ones(len(nonlearn_onsets))*2.5, nonlearn_values)).T            
            remaining_mtrx = np.vstack((remaining_onsets, np.ones(len(remaining_onsets))*2.5, 
                                        np.ones(len(remaining_onsets)))).T
          
            if j == 0: #if the first run in a stimulus set
                np.savetxt(ev_dir + 'run{0}.txt'.format(i*2+1), mtrx, delimiter='\t', fmt='%.4f')                
                for trial in ['learn_event', 'learn_mod', 'nonlearn_event', 'nonlearn_mod', 'remaining']:
                    exec('np.savetxt(ev_dir + "run{0}_{1}.txt", {1}_mtrx,delimiter="\t", fmt="%.4f")'.format(i*2+1,trial))
            else: #if the second run in a stimulus set
                np.savetxt(ev_dir + 'run{0}.txt'.format(i*2+2), mtrx, delimiter='\t', fmt='%.4f')                
                for trial in ['learn_event', 'learn_mod', 'nonlearn_event', 'nonlearn_mod', 'remaining']:
                    exec('np.savetxt(ev_dir + "run{0}_{1}.txt", {1}_mtrx, delimiter="\t", fmt="%.4f")'.format(i*2+2,trial))

subs_list = pd.Series(subs) #convert subs to pandas series
database_df = pd.DataFrame(database) #convert database to pandas dataframe
database_df = database_df.set_index(subs_list) #use new series as row index
```

```python
for sub in subs:
    sub_dir = '/home/data/madlab/Mattfeld_WMAZE/'
    learning_files = sorted(glob(join(sub_dir,'behav/{0}/B_pmode_set*.txt'.format(sub))))     
    upper_95_files = sorted(glob(join(sub_dir, 'behav/{0}/B_p95_set*.txt'.format(sub))))   
    
    for i in range(3):  
        learning_curve = np.loadtxt(learning_files[i])
        deriv_file = (np.gradient(learning_curve)*10) 
        upper_95 = np.loadtxt(upper_95_files[i])
        
        #plt.plot(deriv_file) #before shaving
        run1_deriv = deriv_file[:40]
        run1_learn = learning_curve[:40]
        run1_95 = upper_95[:40]
        #print alt_deriv_file 
        #plt.plot(alt_deriv_file) #after shaving
        #target_trial = np.argmax(run1_deriv)
        run2_deriv = deriv_file[40:]
        run2_learn = learning_curve[40:]
        run2_95 = upper_95[40:]
        #print target_trial
        #plt.plot(learning_curve)
        
        x = range(len(run1_learn))
        chance = np.ones(len(x)) * 0.5
        fig, (ax0, ax1) = plt.subplots(ncols = 2, figsize=(11, 5), sharex = False, sharey = True)  
        ax0.plot(x, run1_learn, "#B266FF", linewidth = 3)
        ax0.plot(x, chance, 'b', linewidth = 1)
        ax0.fill_between(x, run1_learn, run1_95, color = "#B266FF", alpha = 0.2)
        ax0.plot(x, run1_deriv, "#00CCCC", linewidth = 3)
        ax0.set_ylim(0,1)
        ax0.set_xlabel('Trials')
        ax0.set_ylabel('Probability of Correct Response')
        ax0.set_title("Set {0} Run 1".format(i+1))
        
        x1 = range(len(run2_learn))
        chance1 = np.ones(len(x1)) * 0.5
        ax1.plot(x1, run2_learn, "#B266FF", linewidth = 3)
        ax1.plot(x1, chance1, 'b', linewidth = 1)
        ax1.fill_between(x1, run2_learn, run2_95, color = "#B266FF", alpha = 0.2)
        ax1.plot(x1, run2_deriv, "#00CCCC", linewidth = 3)        
        ax1.set_ylim(0,1)
        ax1.set_xlabel('Trials')
        ax1.set_title("Set {0} Run 2".format(i+1))
        
        print sub
        print "Run 1 Learn: ", len(np.where(run1_deriv > 0)[0])
        print "Run 1 Nonlearn: ", len(np.where(run1_deriv <= 0)[0])
        print "Run 2 Learn: ", len(np.where(run2_deriv > 0)[0])
        print "Run 2 Nonlearn: ", len(np.where(run2_deriv <= 0)[0])
        plt.show()
```
