```python
import os
from os.path import join
import numpy as np
import scipy as sp
import nibabel as nb
%matplotlib inline
import matplotlib.pyplot as plt
from nipy.labs.viz_tools.activation_maps import plot_map
from surfer import Brain, io
from IPython.core.display import Image


#Fixed minus conditional
fixed_file = join("/home/data/madlab/data/mri/wmaze/grplvl/grplvl_MRthesis_randomise/comprehensive2/",
                  "allFixed_minus_allCond/output/corrected/_contrast_allFixed_minus_allCond",
                  "oneSampT_tfce_corrp_tstat1.nii.gz")
fixed_nii = nb.load(fixed_file)

plot_map(fixed_nii.get_data(), fixed_nii.get_affine(), threshold = 0.95,
         draw_cross=False, vmin = 0.95, vmax = 1, cmap = 'autumn')
```

```python
#If this cell fails, load emu_env, THEN wmaze_env

reg_file = os.path.join(os.environ["FREESURFER_HOME"], "average/mni152.register.dat")
surf_fixed = io.project_volume_data(fixed_file, "lh", reg_file) #reds
```

## LEFT HEMISPHERE

```python
brain = Brain("/home/data/madlab/surfaces/wmaze/fsaverage/", "lh", "inflated")
#correct fixed minus incorrect fixed
brain.add_overlay(surf_fixed, min = .95, max = 1, name = "surf_fixed")
brain.overlays["surf_fixed"].pos_bar.lut_mode = "Purples"
#produce brain image
brain.save_image("/tmp/my3dbrain.png")
Image("/tmp/my3dbrain.png")
```
