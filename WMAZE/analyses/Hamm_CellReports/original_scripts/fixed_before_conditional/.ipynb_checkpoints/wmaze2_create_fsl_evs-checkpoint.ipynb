```python
import os
from os.path import join
import numpy as np
import pandas as pd
from glob import glob
```

```python
# Array containing subject ids

#subs = ['WMAZE_001']

subs = ['WMAZE_001', 'WMAZE_002', 'WMAZE_003', 'WMAZE_004', 'WMAZE_005',
        'WMAZE_006', 'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010',
        'WMAZE_012', 'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020',
        'WMAZE_021', 'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 
        'WMAZE_027']


# Array containing the three sets
stim_sets = ['set1', 'set2', 'set3']

# Loop to grab correct 6 run text files for each subject
for sub in subs:
    # Array containing path to behavior files
    sub_dir = '/home/data/madlab/data/mri/wmaze/scanner_behav/{0}/'.format(sub)
    # Array containing current sub's 6 behavior file runs
    dir_file = glob(join(sub_dir, '{0}_wmazebl_2015*.txt'.format(sub)))    
    # Sort current sub's txt files in order of run
    dir_file.sort() 

       
    # Loop through each of the set types (set1, set2, set3)
    for i, curr_set in enumerate(stim_sets):
        # Create dataframe for the run 1 & 2 text files to extract EVS
        run1 = pd.read_table(dir_file[i * 2])
        run2 = pd.read_table(dir_file[i * 2 + 1])


        # Convert dataframes into numpy arrays for trial type and correct/incorrect
        # Run 1
        run1_trialtype = run1['TrialType'].values
        run1_correct = run1['Correct'].values
        # Run 2
        run2_trialtype = run2['TrialType'].values
        run2_correct = run2['Correct'].values

        # Shift the TrialType column down by 1
        run1_trial_shift = run1_trialtype[:-1]
        # Insert dummy element into the first index
        run1_trial_shift = np.insert(run1_trial_shift, 0, -1)
        # Shift the TrialType column down by 1
        run2_trial_shift = run2_trialtype[:-1]
        # Insert dummy element into the first index
        run2_trial_shift = np.insert(run2_trial_shift, 0, -1)

        # Shift the Correct column down by 1
        run1_correct_shift = run1_correct[:-1]
        # Insert dummy element into the first index
        run1_correct_shift = np.insert(run1_correct_shift, 0, -1)
        # Shift the Correct column down by 1
        run2_correct_shift = run2_correct[:-1]
        # Insert dummy element into the first index
        run2_correct_shift = np.insert(run2_correct_shift, 0, -1)


        
        
        ## Run #1 ##
        # Grab rows where: A after correct B
        run1_corr_B_before_A = np.where((run1_trial_shift == 'B') * (run1_trialtype == 'A') * (run1_correct_shift == 1))
        # Grab rows where: A after incorrect B
        run1_incorr_B_before_A = np.where((run1_trial_shift == 'B') * (run1_trialtype == 'A') * (run1_correct_shift == 0))
        # Grab rows where: C after correct B
        run1_corr_B_before_C = np.where((run1_trial_shift == 'B') * (run1_trialtype == 'C') * (run1_correct_shift == 1))
        # Grab rows where: C after incorrect B
        run1_incorr_B_before_C = np.where((run1_trial_shift == 'B') * (run1_trialtype == 'C') * (run1_correct_shift == 0))
        # Grab rows where: all trials not preceeded by B or if currenct trial is BL
        run1_all_remaining = np.where((run1_trial_shift != 'B') + (run1_trialtype == 'BL'))


        ## Run #2 ##
        # Grab rows where: A after correct B
        run2_corr_B_before_A = np.where((run2_trial_shift == 'B') * (run2_trialtype == 'A') * (run2_correct_shift == 1))
        # Grab rows where: A after incorrect B
        run2_incorr_B_before_A = np.where((run2_trial_shift == 'B') * (run2_trialtype == 'A') * (run2_correct_shift == 0))
        # Grab rows where: C after correct B
        run2_corr_B_before_C = np.where((run2_trial_shift == 'B') * (run2_trialtype == 'C') * (run2_correct_shift == 1))
        # Grab rows where: C after incorrect B
        run2_incorr_B_before_C = np.where((run2_trial_shift == 'B') * (run2_trialtype == 'C') * (run2_correct_shift == 0))
        # Grab rows where: all trials not preceeded by B or if currenct trial is BL
        run2_all_remaining = np.where((run2_trial_shift != 'B') + (run1_trialtype == 'BL'))


        # All onsets from run 1
        run1_onsets = run1['StimOnset']
        # All onsets from run 2
        run2_onsets = run2['StimOnset']


        # Only A preceeded by correct B onsets from run 1 
        run1_corr_B_before_A_onsets = run1['StimOnset'].values[run1_corr_B_before_A[0]]
        # Only A preceeded by incorrect B onsets from run 1
        run1_incorr_B_before_A_onsets = run1['StimOnset'].values[run1_incorr_B_before_A[0]]
        # Only C preceeded by correct B onsets from run 1
        run1_corr_B_before_C_onsets = run1['StimOnset'].values[run1_corr_B_before_C[0]]
        # Only C preceeded by incorrect B onsets from run 1
        run1_incorr_B_before_C_onsets = run1['StimOnset'].values[run1_incorr_B_before_C[0]]
        # All other trials from run 1
        run1_all_remaining_onsets = run1['StimOnset'].values[run1_all_remaining]

        # Only A preceeded by correct B onsets from run 2 
        run2_corr_B_before_A_onsets = run2['StimOnset'].values[run2_corr_B_before_A[0]]
        # Only A preceeded by incorrect B onsets from run 2
        run2_incorr_B_before_A_onsets = run2['StimOnset'].values[run2_incorr_B_before_A[0]]
        # Only C preceeded by correct B onsets from run 2
        run2_corr_B_before_C_onsets = run2['StimOnset'].values[run2_corr_B_before_C[0]]
        # Only C preceeded by incorrect B onsets from run 2
        run2_incorr_B_before_C_onsets = run2['StimOnset'].values[run2_incorr_B_before_C[0]]
        # All other trials from run 2
        run2_all_remaining_onsets = run2['StimOnset'].values[run2_all_remaining]


        ### Run #1 ###               
        # Use v-stack to create a matrix containing *ALL* onsets, durations, 
        # and amplitudes in vertical columns for run 1
        run1_mtrx = np.vstack((run1_onsets,
                               # Numpy array filled with 3's
                               np.ones(len(run1_onsets)) * 3.0,
                               # Numpy array filled with 1's
                               np.ones(len(run1_onsets)))).T

        # Use v-stack to create a matrix containing *CORRECT B BEFORE A ONLY* onsets, durations, 
        # and amplitudes in vertical columns for run 1            
        run1_corr_B_before_A_mtrx = np.vstack((run1_corr_B_before_A_onsets, 
                                               np.ones(len(run1_corr_B_before_A_onsets)) * 3.0,
                                               np.ones(len(run1_corr_B_before_A_onsets)))).T

        # Use v-stack to create a matrix containing *INCORRECT B BEFORE A ONLY* onsets, durations, 
        # and amplitudes in vertical columns for run 1
        run1_incorr_B_before_A_mtrx = np.vstack((run1_incorr_B_before_A_onsets,
                                                 np.ones(len(run1_incorr_B_before_A_onsets)) * 3.0,
                                                 np.ones(len(run1_incorr_B_before_A_onsets)))).T

        # Use v-stack to create a matrix containing *CORRECT B BEFORE C ONLY* onsets, durations, 
        # and amplitudes in vertical columns for run 1            
        run1_corr_B_before_C_mtrx = np.vstack((run1_corr_B_before_C_onsets,
                                         np.ones(len(run1_corr_B_before_C_onsets)) * 3.0,
                                         np.ones(len(run1_corr_B_before_C_onsets)))).T

        # Use v-stack to create a matrix containing *INCORRECT B BEFORE C ONLY* onsets, durations, 
        # and amplitudes in vertical columns for run 1
        run1_incorr_B_before_C_mtrx = np.vstack((run1_incorr_B_before_C_onsets,
                                            np.ones(len(run1_incorr_B_before_C_onsets)) * 3.0,
                                            np.ones(len(run1_incorr_B_before_C_onsets)))).T

        # Use v-stack to create a matrix containing *ALL REMAINING TRIALS* onsets, durations, 
        # and amplitudes in vertical columns for run 1
        run1_all_remaining_mtrx = np.vstack((run1_all_remaining_onsets,
                                            np.ones(len(run1_all_remaining_onsets)) * 3.0,
                                            np.ones(len(run1_all_remaining_onsets)))).T


        ### RUN #2 ###               
        # Use v-stack to create a matrix containing *ALL* onsets, durations, 
        # and amplitudes in vertical columns for run 2
        run2_mtrx = np.vstack((run2_onsets,
                               # Numpy array filled with 3's
                               np.ones(len(run2_onsets)) * 3.0,
                               # Numpy array filled with 1's
                               np.ones(len(run2_onsets)))).T

        # Use v-stack to create a matrix containing *CORRECT B BEFORE A ONLY* onsets, durations, 
        # and amplitudes in vertical columns for run 2            
        run2_corr_B_before_A_mtrx = np.vstack((run2_corr_B_before_A_onsets,
                                         np.ones(len(run2_corr_B_before_A_onsets)) * 3.0,
                                         np.ones(len(run2_corr_B_before_A_onsets)))).T

        # Use v-stack to create a matrix containing *INCORRECT B BEFORE A ONLY* onsets, durations, 
        # and amplitudes in vertical columns for run 2
        run2_incorr_B_before_A_mtrx = np.vstack((run2_incorr_B_before_A_onsets,
                                            np.ones(len(run2_incorr_B_before_A_onsets)) * 3.0,
                                            np.ones(len(run2_incorr_B_before_A_onsets)))).T

        # Use v-stack to create a matrix containing *CORRECT B BEFORE C ONLY* onsets, durations, 
        # and amplitudes in vertical columns for run 2            
        run2_corr_B_before_C_mtrx = np.vstack((run2_corr_B_before_C_onsets,
                                         np.ones(len(run2_corr_B_before_C_onsets)) * 3.0,
                                         np.ones(len(run2_corr_B_before_C_onsets)))).T

        # Use v-stack to create a matrix containing *INCORRECT B BEFORE C ONLY* onsets, durations, 
        # and amplitudes in vertical columns for run 2
        run2_incorr_B_before_C_mtrx = np.vstack((run2_incorr_B_before_C_onsets,
                                            np.ones(len(run2_incorr_B_before_C_onsets)) * 3.0,
                                            np.ones(len(run2_incorr_B_before_C_onsets)))).T

        # Use v-stack to create a matrix containing *ALL REMAINING TRIALS* onsets, durations, 
        # and amplitudes in vertical columns for run 1
        run2_all_remaining_mtrx = np.vstack((run2_all_remaining_onsets,
                                            np.ones(len(run2_all_remaining_onsets)) * 3.0,
                                            np.ones(len(run2_all_remaining_onsets)))).T


        # If the output directory does not exist
        if not os.path.exists(join(sub_dir, 'MRthesis/', 'cond_b4_fixed/', 'EVs/')):
            # Create it
            os.makedirs(join(sub_dir, 'MRthesis/', 'cond_b4_fixed/', 'EVs/')) 


        ## Run 1 ##    
        # Create EV text file for Set_-Run 1 ALL onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}.txt'.format(i * 2), 
                   run1_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-Run1 *CORRECT B BEFORE A ONLY* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_corr_B_before_A_mtrx.txt'.format(i * 2 + 1), 
                   run1_corr_B_before_A_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-Run1 *CORRECT B BEFORE A ONLY* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_incorr_B_before_A_mtrx.txt'.format(i * 2 + 1), 
                   run1_incorr_B_before_A_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-Run1 *CORRECT B BEFORE C ONLY* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_corr_B_before_C_mtrx.txt'.format(i * 2 + 1), 
                   run1_corr_B_before_C_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-Run1 *CORRECT B BEFORE C ONLY* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_incorr_B_before_C_mtrx.txt'.format(i * 2 + 1), 
                   run1_incorr_B_before_C_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-Run1 *ALL REMAINING* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_all_remaining_mtrx.txt'.format(i * 2 + 1), 
                   run1_all_remaining_mtrx, delimiter='\t', fmt='%.4f')


        ## Run 2 ##
        # Create EV text file for Set_-run2 ALL onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}.txt'.format(i * 2 + 2), 
                   run2_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-run2 *CORRECT B BEFORE A ONLY* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_corr_B_before_A_mtrx.txt'.format(i * 2 + 2), 
                   run2_corr_B_before_A_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-run2 *CORRECT B BEFORE A ONLY* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_incorr_B_before_A_mtrx.txt'.format(i * 2 + 2), 
                   run2_incorr_B_before_A_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-run2 *CORRECT B BEFORE C ONLY* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_corr_B_before_C_mtrx.txt'.format(i * 2 + 2), 
                   run2_corr_B_before_C_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-run2 *CORRECT B BEFORE C ONLY* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_incorr_B_before_C_mtrx.txt'.format(i * 2 + 2), 
                   run2_incorr_B_before_C_mtrx, delimiter='\t', fmt='%.4f')

        # Create EV text file for Set_-run2 *ALL REMAINING* onsets/durations/amplitudes
        np.savetxt(sub_dir + '/MRthesis/' + 'cond_b4_fixed/' + 'EVs/' + 
                   'run{0}_all_remaining_mtrx.txt'.format(i * 2 + 2), 
                   run2_all_remaining_mtrx, delimiter='\t', fmt='%.4f')             
```
