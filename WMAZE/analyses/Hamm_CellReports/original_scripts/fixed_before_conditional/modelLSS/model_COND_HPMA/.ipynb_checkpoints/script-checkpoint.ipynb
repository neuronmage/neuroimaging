```python
import os
from os.path import join
from glob import glob
from scipy.stats import pearsonr
from pylab import *
import numpy as np
import nibabel as nb
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette("husl")
```

```python
sids = ['WMAZE_001', 'WMAZE_002', 'WMAZE_004', 'WMAZE_005', 'WMAZE_006',
        'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 'WMAZE_012',
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021',
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']


proj_dir = '/home/data/madlab/data/mri/wmaze' 
mask_filenames = []
cope_files = []
for SID in sids:
    mask_filenames_glob = glob(proj_dir + '/roi_analysis/MRthesis/fb4c_2/mask/anat_masks/_subject_id_' 
                               + SID + '/_anatmask_xfm*/*')
    mask_filenames.append(mask_filenames_glob)
    subjcopes_glob = glob(proj_dir + '/scndlvl/wmaze_MRthesis/fixed_before_conditional/model_HALVES/' 
                          + SID + '/fixedfx/cope_curve*')
    subjcopes_glob.sort()
    cope_files.append(subjcopes_glob)
    
    if len(cope_files[-1]) == 0:
        print(SID)
```

```python
all_data = {'subjid':[],
            'run1_RSC_A':[], 'run1_RSC_B':[], 'run1_RSC_C':[],
            'run1_HPC_A':[], 'run1_HPC_B':[], 'run1_HPC_C':[],
            'run1_VV_A':[], 'run1_VV_B':[], 'run1_VV_C':[],
            'run1_dmPFC_A':[], 'run1_dmPFC_B':[], 'run1_dmPFC_C':[],
            'run1_dlPFC_A':[], 'run1_dlPFC_B':[], 'run1_dlPFC_C':[],
            
            'run2_RSC_A':[], 'run2_RSC_B':[], 'run2_RSC_C':[],
            'run2_HPC_A':[], 'run2_HPC_B':[], 'run2_HPC_C':[],
            'run2_VV_A':[], 'run2_VV_B':[], 'run2_VV_C':[],
            'run2_dmPFC_A':[], 'run2_dmPFC_B':[], 'run2_dmPFC_C':[],
            'run2_dlPFC_A':[], 'run2_dlPFC_B':[], 'run2_dlPFC_C':[]}

for i in range(len(subs)):
    all_data['subjid'].append(subs[i])
    RSC_img = nb.load(mask_filenames[i][0])
    HPC_img = nb.load(mask_filenames[i][1])
    VV_img = nb.load(mask_filenames[i][2])
    dmPFC_img = nb.load(mask_filenames[i][3])
    dlPFC_img = nb.load(mask_filenames[i][4])
    
    run1_A_img = nb.load(cope_files[i][0])
    run1_B_img = nb.load(cope_files[i][1])
    run1_C_img = nb.load(cope_files[i][2])
    run2_A_img = nb.load(cope_files[i][3])
    run2_B_img = nb.load(cope_files[i][4])
    run2_C_img = nb.load(cope_files[i][5])
    
    region = ['HPC', 'VV', 'dmPFC', 'dlPFC']
    stimulus = ['A', 'B', 'C']
   
    for r in region:
        for s in stimulus:
            run1_data = eval('run1_{0}_img.get_data()[{1}_img.get_data() > 0.]'.format(s,r))
            all_data['run1_{0}_{1}'.format(r,s)].append(np.mean(run1_data))            
            run2_data = eval('run2_{0}_img.get_data()[{1}_img.get_data() > 0.]'.format(s,r))
            all_data['run2_{0}_{1}'.format(r,s)].append(np.mean(run2_data))
    
    for s in stimulus:
        run1_data = eval('run1_{0}_img.get_data()[RSC_img.get_data()[:,:,:,0] > 0.]'.format(s))
        all_data['run1_RSC_{0}'.format(s)].append(np.mean(run1_data))            
        run2_data = eval('run2_{0}_img.get_data()[RSC_img.get_data()[:,:,:,0] > 0.]'.format(s))
        all_data['run2_RSC_{0}'.format(s)].append(np.mean(run2_data))

all_data_df = pd.DataFrame(all_data)
```

```python
region = ['RSC', 'HPC', 'VV', 'dmPFC', 'dlPFC']
stimulus = ['A', 'B', 'C']
plot = {}

for r in region:
    for s in stimulus:
        all_data_df['allrun_{0}_{1}'.format(r,s)] = (all_data_df['run1_{0}_{1}'.format(r,s)] 
                                                     + all_data_df['run1_{0}_{1}'.format(r,s)])/2.
        plot['{0}_{1}'.format(r,s)] = all_data_df['allrun_{0}_{1}'.format(r,s)]
plot_df = pd.DataFrame.from_dict(plot)
```

```python
subs = ['WMAZE_001', 'WMAZE_002', 'WMAZE_004', 'WMAZE_005', 'WMAZE_006', 
        'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 'WMAZE_012', 
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021', 
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']

stim_dict = {'all_prop_A':[], 'all_prop_B':[], 'all_prop_C':[]}

for sub in subs:
    sub_dir = '/home/data/madlab/data/mri/wmaze/scanner_behav/{0}/'.format(sub)
    dir_file = glob(join(sub_dir, '{0}_wmazebl_2015*.txt'.format(sub)))   
    dir_file.sort()    
    
    data_run1 = pd.read_table(dir_file[0])
    data_run1 = data_run1[:-3]
    data_run2 = pd.read_table(dir_file[1])
    data_run2 = data_run2[:-3]
    data_run3 = pd.read_table(dir_file[2])
    data_run3 = data_run3[:-3]
    data_run4 = pd.read_table(dir_file[3])
    data_run4 = data_run4[:-3]
    data_run5 = pd.read_table(dir_file[4])
    data_run5 = data_run5[:-3]
    data_run6 = pd.read_table(dir_file[5])
    data_run6 = data_run6[:-3]
    
    all_runs = [data_run1, data_run2, data_run3, data_run4, data_run5, data_run6]    
    data_set = pd.concat(all_runs)
    
    trialtype = data_set['TrialType'].values
    correct = data_set['Correct'].values 
    
    all_A_corr = len(np.where((trialtype == 'A') & (correct == 1))[0])
    all_A_incorr = len(np.where((trialtype == 'A') & (correct == 0))[0])
    all_B_corr = len(np.where((trialtype == 'B') & (correct == 1))[0])
    all_B_incorr = len(np.where((trialtype == 'B') & (correct == 0))[0])
    all_C_corr = len(np.where((trialtype == 'C') & (correct == 1))[0])
    all_C_incorr = len(np.where((trialtype == 'C') & (correct == 0))[0])
    
    stim_dict['all_prop_A'].append(all_A_corr/(float(all_A_corr) + all_A_incorr))
    stim_dict['all_prop_B'].append(all_B_corr/(float(all_B_corr) + all_B_incorr))
    stim_dict['all_prop_C'].append(all_C_corr/(float(all_C_corr) + all_C_incorr))

plot_df['prop_A'] = stim_dict['all_prop_A']
plot_df['prop_B'] = stim_dict['all_prop_B']
plot_df['prop_C'] = stim_dict['all_prop_C']
```

```python
N = 15
conditions = ['RSCA', 'RSCB', 'RSCC', 'HPCA', 'HPCB', 'HPCC', 'VVA', 'VVB', 'VVC',
              'dmPFCA', 'dmPFCB', 'dmPFCC', 'dlPFCA', 'dlPFCB', 'dlPFCC'] 

hp_allsubjs = [plot_df['RSC_A'], plot_df['RSC_B'], plot_df['RSC_C'], 
               plot_df['HPC_A'], plot_df['HPC_B'], plot_df['HPC_C'],
               plot_df['VV_A'], plot_df['VV_B'], plot_df['VV_C'],
               plot_df['dmPFC_A'], plot_df['dmPFC_B'], plot_df['dmPFC_C'],
               plot_df['dlPFC_A'], plot_df['dlPFC_B'], plot_df['dlPFC_C']]

ind = np.arange(N)
fig, ax = plt.subplots(figsize= (12,6))
ax0 = sns.boxplot(data = hp_allsubjs, width = 0.3)
ax2 = sns.swarmplot(data = hp_allsubjs, color='.25')
ax.set_xticklabels(conditions)
ax.set_xticks(ind)
ax.set_xlabel("Stimulus Type")
ax.set_ylabel("Arbitrary units")
ax.set_title("Regional Activation")
plt.show()
```
