```python
import os
from os.path import join
from glob import glob
from math import sqrt
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette('muted')
sns.set_style("whitegrid")

#subs = ['WMAZE_001']

subs = ['WMAZE_001', 'WMAZE_002', 'WMAZE_004', 'WMAZE_005', 'WMAZE_006',
        'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 'WMAZE_012',
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021',
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']

stim_sets = ['set1', 'set2', 'set3']

count_table = {}
ctstd_table = {}
rt_table = {}
rtstd_table = {}

for sub in subs:
    ct_dict = {}
    rt_dict = {}
    sub_dir = '/home/data/madlab/data/mri/wmaze/scanner_behav/{0}/'.format(sub)
    dir_file = glob(join(sub_dir, '{0}_wmazebl_2015*.txt'.format(sub)))    
    dir_file.sort() 

    for i, curr_set in enumerate(stim_sets):
        run1 = pd.read_table(dir_file[i * 2])
        run2 = pd.read_table(dir_file[i * 2 + 1])
        
        trialtype = np.empty(320, dtype=object)
        trialtype[:160] = run1['TrialType'].values
        trialtype[160:] = run2['TrialType'].values
        
        trial_shift1 = trialtype[1:] 
        trial_shift1 = np.append(trial_shift1, -1)
        trial_shift2 = trial_shift1[1:] 
        trial_shift2 = np.append(trial_shift2, -1)
        trial_shift3 = trial_shift2[1:] 
        trial_shift3 = np.append(trial_shift3, -1)
        trial_shift4 = trial_shift3[1:] 
        trial_shift4 = np.append(trial_shift4, -1)
        trial_shift5 = trial_shift4[1:] 
        trial_shift5 = np.append(trial_shift5, -1)
           
       
        f_BL0_C = np.where(((trialtype == 'A') | (trialtype == 'C')) & (trial_shift1 == 'B'))[0]

        f_BL1_C = np.where(((trialtype == 'A') | (trialtype == 'C')) & (trial_shift1 == 'BL')
                            & (trial_shift2 == 'B'))[0] 
        f_BL2_C = np.where(((trialtype == 'A') | (trialtype == 'C')) & (trial_shift1 == 'BL')
                            & (trial_shift2 == 'BL') & (trial_shift3 == 'B'))[0] 
        f_BL3_C = np.where(((trialtype == 'A') | (trialtype == 'C')) & (trial_shift1 == 'BL')
                            & (trial_shift2 == 'BL') & (trial_shift3 == 'BL') & (trial_shift4 == 'B'))[0] 
        f_BL4_C = np.where(((trialtype == 'A') | (trialtype == 'C')) & (trial_shift1 == 'BL')
                            & (trial_shift2 == 'BL') & (trial_shift3 == 'BL') & (trial_shift4 == 'BL')
                            & (trial_shift5 == 'B'))[0] 
        
         
        for curr_type in ['f_BL0_C', 'f_BL1_C', 'f_BL2_C', 'f_BL3_C', 'f_BL4_C']:
            curr_name = '{0}'.format(curr_type)
            if not curr_name in ct_dict:
                ct_dict[curr_name] = []
            ct_dict[curr_name].append(len(eval(curr_name)))                          
    
    for key in ct_dict:
        ct_dict[key] = np.sum(ct_dict[key])
        if not key in count_table:
            count_table[key] = []
        count_table[key].append(ct_dict[key])
                
BLs_df = pd.DataFrame(count_table, index = subs)  
```

```python
ct_avg = {}
ct_std = {}
for curr_key in count_table:
    ct_avg[curr_key] = np.average(count_table[curr_key])
    ct_std[curr_key] = np.std(count_table[curr_key])    
count_average = pd.DataFrame(ct_avg, index = (1,))
count_std = pd.DataFrame(ct_std, index = (1,))
```

```python
import os
from os.path import join
from glob import glob
from math import sqrt
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette('muted')
sns.set_style("whitegrid")

#subs = ['WMAZE_001']

subs = ['WMAZE_001', 'WMAZE_002', 'WMAZE_004', 'WMAZE_005', 'WMAZE_006',
        'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 'WMAZE_012',
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021',
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']

stim_sets = ['set1', 'set2', 'set3']

count_table = {}
ctstd_table = {}
rt_table = {}
rtstd_table = {}

for sub in subs:
    ct_dict = {}
    rt_dict = {}
    sub_dir = '/home/data/madlab/data/mri/wmaze/scanner_behav/{0}/'.format(sub)
    dir_file = glob(join(sub_dir, '{0}_wmazebl_2015*.txt'.format(sub)))    
    dir_file.sort() 

    for i, curr_set in enumerate(stim_sets):
        run1 = pd.read_table(dir_file[i * 2])
        run2 = pd.read_table(dir_file[i * 2 + 1])
        
        run1 = run1[:-3]
        run2 = run2[:-3]
        
        run1_trialtype = run1['TrialType'].values
        run1_correct = run1['Correct'].values
        run1_resp = run1['Resp'].values 
        run2_trialtype = run2['TrialType'].values
        run2_correct = run2['Correct'].values
        run2_resp = run2['Resp'].values
        #print run1_trialtype
        
        run1_fixed_before = run1_trialtype[:-1]
        run1_fixed_before = np.insert(run1_fixed_before, 0, -1)
        run1_trial_shift1 = run1_trialtype[1:] 
        run1_trial_shift1 = np.append(run1_trial_shift1, -1)
        run1_trial_shift2 = run1_trial_shift1[1:] 
        run1_trial_shift2 = np.append(run1_trial_shift2, -1)        
        run2_fixed_before = run2_trialtype[:-1]
        run2_fixed_before = np.insert(run2_fixed_before, 0, -1)
        run2_trial_shift1 = run2_trialtype[1:] 
        run2_trial_shift1 = np.append(run2_trial_shift1, -1)
        run2_trial_shift2 = run2_trial_shift1[1:] 
        run2_trial_shift2 = np.append(run2_trial_shift2, -1)
        

        run1_correct_shift1 = run1_correct[1:] 
        run1_correct_shift1 = np.append(run1_correct_shift1, -1)
        run1_correct_shift2 = run1_correct_shift1[1:] 
        run1_correct_shift2 = np.append(run1_correct_shift2, -1) 
        run2_correct_shift1 = run2_correct[1:] 
        run2_correct_shift1 = np.append(run2_correct_shift1, -1)
        run2_correct_shift2 = run2_correct_shift1[1:] 
        run2_correct_shift2 = np.append(run2_correct_shift2, -1)
           
       
        run1_f_C_corr = np.where((run1_trialtype != 'BL') & (run1_trial_shift1 == 'B') & (run1_correct_shift1 == 1))[0]       
        run1_f_C_incorr = np.where((run1_trialtype != 'BL') & (run1_trial_shift1 == 'B') & (run1_correct_shift1 == 0))[0]         
        run1_f_BL1_C = np.where(((run1_fixed_before == 'A') | (run1_fixed_before == 'C')) 
                                  & (run1_trialtype == 'BL') & (run1_trial_shift1 == 'B'))[0]         
        temp1 = np.concatenate([run1_f_C_corr, run1_f_C_incorr, run1_f_BL1_C])
        temp1 = set(temp1)
        x1 = set(np.arange(0,157))
        run1_all_remaining = x1.difference(temp1)


        run2_f_C_corr = np.where((run2_trialtype != 'BL') & (run2_trial_shift1 == 'B') & (run2_correct_shift1 == 1))[0]
        run2_f_C_incorr = np.where((run2_trialtype != 'BL') & (run2_trial_shift1 == 'B') & (run2_correct_shift1 == 0))[0]      
        run2_f_BL1_C = np.where(((run2_fixed_before == 'A') | (run2_fixed_before == 'C')) 
                                  & (run2_trialtype == 'BL') & (run2_trial_shift1 == 'B'))[0]            
        temp2 = np.concatenate([run2_f_C_corr, run2_f_C_incorr, run2_f_BL1_C])
        temp2 = set(temp2)
        x2 = set(np.arange(0,157))
        run2_all_remaining = x2.difference(temp2)
        
        # All onsets
        run1_onsets = run1['StimOnset']
        run2_onsets = run2['StimOnset']        
 
        run1_f_C_corr_onsets = run1_onsets.values[run1_f_C_corr]
        run1_f_C_incorr_onsets = run1_onsets.values[run1_f_C_incorr]
        run1_f_BL1_C_onsets = run1_onsets.values[run1_f_BL1_C]
        run1_all_remaining_onsets = run1_onsets.values[list(run1_all_remaining)]
        
        run2_f_C_corr_onsets = run2_onsets.values[run2_f_C_corr]
        run2_f_C_incorr_onsets = run2_onsets.values[run2_f_C_incorr]
        run2_f_BL1_C_onsets = run2_onsets.values[run2_f_BL1_C]
        run2_all_remaining_onsets = run2_onsets.values[list(run2_all_remaining)]
        

        run1_mtrx = np.vstack((run1_onsets,
                               np.ones(len(run1_onsets)) * 3.0,
                               np.ones(len(run1_onsets)))).T

        run1_f_C_corr_mtrx = np.vstack((run1_f_C_corr_onsets, 
                                        np.ones(len(run1_f_C_corr_onsets)) * 3.0,
                                        np.ones(len(run1_f_C_corr_onsets)))).T
 
        run1_f_C_incorr_mtrx = np.vstack((run1_f_C_incorr_onsets,
                                          np.ones(len(run1_f_C_incorr_onsets)) * 3.0,
                                          np.ones(len(run1_f_C_incorr_onsets)))).T
    
        run1_f_BL1_C_mtrx = np.vstack((run1_f_BL1_C_onsets, 
                                            np.ones(len(run1_f_BL1_C_onsets)) * 3.0,
                                            np.ones(len(run1_f_BL1_C_onsets)))).T


        run1_all_remaining_mtrx = np.vstack((run1_all_remaining_onsets,
                                             np.ones(len(run1_all_remaining_onsets)) * 3.0,
                                             np.ones(len(run1_all_remaining_onsets)))).T

        run2_mtrx = np.vstack((run2_onsets,
                               np.ones(len(run2_onsets)) * 3.0,
                               np.ones(len(run2_onsets)))).T

        run2_f_C_corr_mtrx = np.vstack((run2_f_C_corr_onsets, 
                                        np.ones(len(run2_f_C_corr_onsets)) * 3.0,
                                        np.ones(len(run2_f_C_corr_onsets)))).T
 
        run2_f_C_incorr_mtrx = np.vstack((run2_f_C_incorr_onsets,
                                          np.ones(len(run2_f_C_incorr_onsets)) * 3.0,
                                          np.ones(len(run2_f_C_incorr_onsets)))).T
    
        run2_f_BL1_C_mtrx = np.vstack((run2_f_BL1_C_onsets, 
                                       np.ones(len(run2_f_BL1_C_onsets)) * 3.0,
                                       np.ones(len(run2_f_BL1_C_onsets)))).T

        run2_all_remaining_mtrx = np.vstack((run2_all_remaining_onsets,
                                             np.ones(len(run2_all_remaining_onsets)) * 3.0,
                                             np.ones(len(run2_all_remaining_onsets)))).T

        if not os.path.exists(join(sub_dir, 'MRthesis/', 'model_REVIEW/', 'EVs/')):
            os.makedirs(join(sub_dir, 'MRthesis/', 'model_REVIEW/', 'EVs/')) 
    
        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}.txt'.format(i * 2 + 1), 
                   run1_mtrx, delimiter='\t', fmt='%.4f')

        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}_f_C_corr.txt'.format(i * 2 + 1), 
                   run1_f_C_corr_mtrx, delimiter='\t', fmt='%.4f')

        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}_f_C_incorr.txt'.format(i * 2 + 1), 
                   run1_f_C_incorr_mtrx, delimiter='\t', fmt='%.4f')

        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}_f_BL1_C.txt'.format(i * 2 + 1), 
                   run1_f_BL1_C_mtrx, delimiter='\t', fmt='%.4f')
        
        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}_all_remaining.txt'.format(i * 2 + 1), 
                   run1_all_remaining_mtrx, delimiter='\t', fmt='%.4f')
        
        
        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}.txt'.format(i * 2 + 2), 
                   run2_mtrx, delimiter='\t', fmt='%.4f')

        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}_f_C_corr.txt'.format(i * 2 + 2), 
                   run2_f_C_corr_mtrx, delimiter='\t', fmt='%.4f')

        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}_f_C_incorr.txt'.format(i * 2 + 2), 
                   run2_f_C_incorr_mtrx, delimiter='\t', fmt='%.4f')

        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}_f_BL1_C.txt'.format(i * 2 + 2), 
                   run2_f_BL1_C_mtrx, delimiter='\t', fmt='%.4f')
        
        np.savetxt(sub_dir + '/MRthesis/' + 'model_REVIEW/' + 'EVs/' + 'run{0}_all_remaining.txt'.format(i * 2 + 2), 
                   run2_all_remaining_mtrx, delimiter='\t', fmt='%.4f')          
```
