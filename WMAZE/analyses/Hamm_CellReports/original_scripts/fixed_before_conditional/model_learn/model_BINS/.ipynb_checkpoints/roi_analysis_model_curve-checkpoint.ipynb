# Model Learn Curve

### Model Learning Curve -- Compares learning bins based on the learning curve

```python
from glob import glob
from scipy import stats
from math import sqrt
import nibabel as nb
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_palette('muted')


sids = ['WMAZE_001', 'WMAZE_002', 'WMAZE_004', 'WMAZE_005', 'WMAZE_006',
        'WMAZE_007', 'WMAZE_008', 'WMAZE_009', 'WMAZE_010', 'WMAZE_012',
        'WMAZE_017', 'WMAZE_018', 'WMAZE_019', 'WMAZE_020', 'WMAZE_021',
        'WMAZE_022', 'WMAZE_023', 'WMAZE_024', 'WMAZE_026', 'WMAZE_027']


proj_dir = '/home/data/madlab/data/mri/wmaze' 
mask_filenames = []
cope_files = []
for SID in sids:
    mask_filenames_glob = glob(proj_dir + '/roi_analysis/MRthesis/fb4c_2/mask/anat_masks/_subject_id_' 
                               + SID + '/_anatmask_xfm*/*')
    mask_filenames.append(mask_filenames_glob)
    subjcopes_glob = glob(proj_dir + '/scndlvl/wmaze_MRthesis/fixed_before_conditional/model_curve/' 
                          + SID + '/fixedfx/cope_curve*')
    subjcopes_glob.sort()
    cope_files.append(subjcopes_glob)
    
    if len(cope_files[-1]) == 0:
        print(SID)
```

## Cell to double-check the array indexing

```python
'''
print '## left hippocampus:', mask_filenames[0][0]
print '## left caudate:', mask_filenames[0][1]
print '## right caudate:', mask_filenames[0][2]
print '## left dmPFC:', mask_filenames[0][3]
print '## right hippocampus:', mask_filenames[0][4]
print '## right dmPFC:', mask_filenames[0][5]
print '## left putamen', mask_filenames[0][6]
print '## right_putamen', mask_filenames[0][7]
'''

print '-----'
print '## Fixed_Bin #1:', cope_files[0][0]
print '## Fixed_Bin #2:', cope_files[0][1]
print '## Fixed_Bin #3:', cope_files[0][2]
print '## Fixed_Bin #4:', cope_files[0][3]
```

### Use binarized mask to obtain activation in left & right hemisphere for each region of interest

```python
all_data = {'subjid':[],
            #HPC
            'lhhp_learning_bin1':[], 'rhhp_learning_bin1':[], 
            'lhhp_learning_bin2':[], 'rhhp_learning_bin2':[],
            'lhhp_learning_bin3':[], 'rhhp_learning_bin3':[],
            'lhhp_learning_bin4':[], 'rhhp_learning_bin4':[],
            #Caudate
            'lhcaud_learning_bin1':[], 'rhcaud_learning_bin1':[], 
            'lhcaud_learning_bin2':[], 'rhcaud_learning_bin2':[],
            'lhcaud_learning_bin3':[], 'rhcaud_learning_bin3':[],
            'lhcaud_learning_bin4':[], 'rhcaud_learning_bin4':[],
            #Putamen
            'lhput_learning_bin1':[], 'rhput_learning_bin1':[], 
            'lhput_learning_bin2':[], 'rhput_learning_bin2':[],
            'lhput_learning_bin3':[], 'rhput_learning_bin3':[],
            'lhput_learning_bin4':[], 'rhput_learning_bin4':[]}

for i in range(len(sids)):
    all_data['subjid'].append(sids[i])
    lh_hp_img = nb.load(mask_filenames[i][0])
    rh_hp_img = nb.load(mask_filenames[i][4])
    lh_caud_img = nb.load(mask_filenames[i][1])
    rh_caud_img = nb.load(mask_filenames[i][2])
    lh_put_img = nb.load(mask_filenames[i][6])
    rh_put_img = nb.load(mask_filenames[i][7])
    
    learning_bin1_img = nb.load(cope_files[i][0])
    learning_bin2_img = nb.load(cope_files[i][1])
    learning_bin3_img = nb.load(cope_files[i][2])
    learning_bin4_img = nb.load(cope_files[i][3])
    
    ########################
    ### Left Hippocampus ###
    ########################
    
    # learning_bin1
    lh_hp_learning_bin1_data = learning_bin1_img.get_data()[lh_hp_img.get_data() > 0.]
    all_data['lhhp_learning_bin1'].append(np.mean(lh_hp_learning_bin1_data))
    
    # learning_bin2
    lh_hp_learning_bin2_data = learning_bin2_img.get_data()[lh_hp_img.get_data() > 0.]
    all_data['lhhp_learning_bin2'].append(np.mean(lh_hp_learning_bin2_data))
    
    # learning_bin3
    lh_hp_learning_bin3_data = learning_bin3_img.get_data()[lh_hp_img.get_data() > 0.]
    all_data['lhhp_learning_bin3'].append(np.mean(lh_hp_learning_bin3_data))
    
    # learning_bin4
    lh_hp_learning_bin4_data = learning_bin4_img.get_data()[lh_hp_img.get_data() > 0.]
    all_data['lhhp_learning_bin4'].append(np.mean(lh_hp_learning_bin4_data))

    
    #########################
    ### Right Hippocampus ###
    #########################
    
    # learning_bin1
    rh_hp_learning_bin1_data = learning_bin1_img.get_data()[rh_hp_img.get_data() > 0.]
    all_data['rhhp_learning_bin1'].append(np.mean(rh_hp_learning_bin1_data))
    
    # learning_bin2
    rh_hp_learning_bin2_data = learning_bin2_img.get_data()[rh_hp_img.get_data() > 0.]
    all_data['rhhp_learning_bin2'].append(np.mean(rh_hp_learning_bin2_data))
    
    # learning_bin3
    rh_hp_learning_bin3_data = learning_bin3_img.get_data()[rh_hp_img.get_data() > 0.]
    all_data['rhhp_learning_bin3'].append(np.mean(rh_hp_learning_bin3_data))

    # learning_bin4
    rh_hp_learning_bin4_data = learning_bin4_img.get_data()[rh_hp_img.get_data() > 0.]
    all_data['rhhp_learning_bin4'].append(np.mean(rh_hp_learning_bin4_data))

  
    ####################
    ### Left Caudate ###
    ####################
    
    # learning_bin1
    lh_caud_learning_bin1_data = learning_bin1_img.get_data()[lh_caud_img.get_data() > 0.]
    all_data['lhcaud_learning_bin1'].append(np.mean(lh_caud_learning_bin1_data))
    
    # learning_bin2
    lh_caud_learning_bin2_data = learning_bin2_img.get_data()[lh_caud_img.get_data() > 0.]
    all_data['lhcaud_learning_bin2'].append(np.mean(lh_caud_learning_bin2_data))
    
    # learning_bin3
    lh_caud_learning_bin3_data = learning_bin3_img.get_data()[lh_caud_img.get_data() > 0.]
    all_data['lhcaud_learning_bin3'].append(np.mean(lh_caud_learning_bin3_data)) 
    
    # learning_bin4
    lh_caud_learning_bin4_data = learning_bin4_img.get_data()[lh_caud_img.get_data() > 0.]
    all_data['lhcaud_learning_bin4'].append(np.mean(lh_caud_learning_bin4_data)) 
    
    
    #####################
    ### Right Caudate ###
    ##################### 
    
    # learning_bin1
    rh_caud_learning_bin1_data = learning_bin1_img.get_data()[rh_caud_img.get_data() > 0.]
    all_data['rhcaud_learning_bin1'].append(np.mean(rh_caud_learning_bin1_data))
    
    # learning_bin2
    rh_caud_learning_bin2_data = learning_bin2_img.get_data()[rh_caud_img.get_data() > 0.]
    all_data['rhcaud_learning_bin2'].append(np.mean(rh_caud_learning_bin2_data))
    
    # learning_bin3
    rh_caud_learning_bin3_data = learning_bin3_img.get_data()[rh_caud_img.get_data() > 0.]
    all_data['rhcaud_learning_bin3'].append(np.mean(rh_caud_learning_bin3_data))
    
    # learning_bin4
    rh_caud_learning_bin4_data = learning_bin4_img.get_data()[rh_caud_img.get_data() > 0.]
    all_data['rhcaud_learning_bin4'].append(np.mean(rh_caud_learning_bin4_data))
    
    
    ####################
    ### Left Putamen ###
    ####################
    
     # learning_bin1
    lh_put_learning_bin1_data = learning_bin1_img.get_data()[lh_put_img.get_data() > 0.]
    all_data['lhput_learning_bin1'].append(np.mean(lh_put_learning_bin1_data))
    
    # learning_bin2
    lh_put_learning_bin2_data = learning_bin2_img.get_data()[lh_put_img.get_data() > 0.]
    all_data['lhput_learning_bin2'].append(np.mean(lh_put_learning_bin2_data))
    
    # learning_bin3
    lh_put_learning_bin3_data = learning_bin3_img.get_data()[lh_put_img.get_data() > 0.]
    all_data['lhput_learning_bin3'].append(np.mean(lh_put_learning_bin3_data))

    # learning_bin4
    lh_put_learning_bin4_data = learning_bin4_img.get_data()[lh_put_img.get_data() > 0.]
    all_data['lhput_learning_bin4'].append(np.mean(lh_put_learning_bin4_data))

    
    #####################
    ### Right Putamen ###
    #####################
    
     # learning_bin1
    rh_put_learning_bin1_data = learning_bin1_img.get_data()[rh_put_img.get_data() > 0.]
    all_data['rhput_learning_bin1'].append(np.mean(rh_put_learning_bin1_data))
    
    # learning_bin2
    rh_put_learning_bin2_data = learning_bin2_img.get_data()[rh_put_img.get_data() > 0.]
    all_data['rhput_learning_bin2'].append(np.mean(rh_put_learning_bin2_data))
    
    # learning_bin3
    rh_put_learning_bin3_data = learning_bin3_img.get_data()[rh_put_img.get_data() > 0.]
    all_data['rhput_learning_bin3'].append(np.mean(rh_put_learning_bin3_data)) 

    # learning_bin4
    rh_put_learning_bin4_data = learning_bin4_img.get_data()[rh_put_img.get_data() > 0.]
    all_data['rhput_learning_bin4'].append(np.mean(rh_put_learning_bin4_data)) 

    
all_data_df = pd.DataFrame(all_data)
```

### Combine the two hemispheres to create one regional average

```python
all_data_df['hp_all_learning_bin1'] = (all_data_df['lhhp_learning_bin1'] 
                                       + all_data_df['rhhp_learning_bin1'])/2.
all_data_df['hp_all_learning_bin2'] = (all_data_df['lhhp_learning_bin2']
                                       + all_data_df['rhhp_learning_bin2'])/2.
all_data_df['hp_all_learning_bin3'] = (all_data_df['lhhp_learning_bin3']
                                       + all_data_df['rhhp_learning_bin3'])/2.
all_data_df['hp_all_learning_bin4'] = (all_data_df['lhhp_learning_bin4']
                                       + all_data_df['rhhp_learning_bin4'])/2.


all_data_df['caud_all_learning_bin1'] = (all_data_df['lhcaud_learning_bin1'] 
                                         + all_data_df['rhcaud_learning_bin1'])/2.
all_data_df['caud_all_learning_bin2'] = (all_data_df['lhcaud_learning_bin2']
                                         + all_data_df['rhcaud_learning_bin2'])/2.
all_data_df['caud_all_learning_bin3'] = (all_data_df['lhcaud_learning_bin3'] 
                                         + all_data_df['rhcaud_learning_bin3'])/2.
all_data_df['caud_all_learning_bin4'] = (all_data_df['lhcaud_learning_bin4'] 
                                         + all_data_df['rhcaud_learning_bin4'])/2.


all_data_df['put_all_learning_bin1'] = (all_data_df['lhput_learning_bin1']
                                        + all_data_df['rhput_learning_bin1'])/2.
all_data_df['put_all_learning_bin2'] = (all_data_df['lhput_learning_bin2'] 
                                        + all_data_df['rhput_learning_bin2'])/2.
all_data_df['put_all_learning_bin3'] = (all_data_df['lhput_learning_bin3'] 
                                        +  all_data_df['rhput_learning_bin3'])/2.
all_data_df['put_all_learning_bin4'] = (all_data_df['lhput_learning_bin4'] 
                                        +  all_data_df['rhput_learning_bin4'])/2.
```

```python
hpc_chart = {}
hpc_chart['1_HPC-learning_bin1'] = all_data_df['hp_all_learning_bin1'] 
hpc_chart['2_HPC-learning_bin2'] = all_data_df['hp_all_learning_bin2'] 
hpc_chart['3_HPC-learning_bin3'] = all_data_df['hp_all_learning_bin3']
hpc_chart['4_HPC-learning_bin4'] = all_data_df['hp_all_learning_bin4']
hpc_chart_df = pd.DataFrame.from_dict(hpc_chart)
```

## Hippocampus Activation

```python
print 'bin1 mean:', np.mean(all_data_df['hp_all_learning_bin1'])
print 'bin1 std:', np.std(all_data_df['hp_all_learning_bin1'])
print ""
print 'bin2 mean:', np.mean(all_data_df['hp_all_learning_bin2'])
print 'bin2 std:', np.std(all_data_df['hp_all_learning_bin2'])
print ""
print 'bin3 mean:', np.mean(all_data_df['hp_all_learning_bin3'])
print 'bin3 std:', np.std(all_data_df['hp_all_learning_bin3'])
print ""
print 'bin4 mean:', np.mean(all_data_df['hp_all_learning_bin4'])
print 'bin4 std:', np.std(all_data_df['hp_all_learning_bin4'])
print ""
print "Bin 4 vs. Bin 1: ", stats.ttest_rel(all_data_df['hp_all_learning_bin4'], 
                                           all_data_df['hp_all_learning_bin1'])[:], "*trending*"

N = 4
conditions = ['Learning Bin1', 'Learning Bin2', 'Learning Bin3', 'Learning Bin4']
hp_allsubjs = [all_data_df['hp_all_learning_bin1'], 
               all_data_df['hp_all_learning_bin2'],
               all_data_df['hp_all_learning_bin3'],
               all_data_df['hp_all_learning_bin4']]
ind = np.arange(N)
width = 0.5
fig, ax = plt.subplots(figsize = (10,5))
ax0 = sns.boxplot(data = hp_allsubjs, color = "#278fea", width = 0.3)
ax2 = sns.swarmplot(data = hp_allsubjs, color='.25')
ax.set_xticks(ind)
ax.set_xticklabels(conditions)
ax.set_ylabel("Arbitrary units")
ax.set_title("HPC Activation")
plt.show()
```
